```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head(1)
```

```python
#Merging both the dataframes on basis of 'title' col

movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#Missing data 
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
#Bringing genres, keywords, cast, crew cols into appropriate data format 

#1. Starting with genres
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
#2. Keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#3 Cast
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
#4 Crew 
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
#Now for overview 
# as overview is in string we will convert it into list to concatenate it with other list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#Applying Transformation
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace (" ", "") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace (" ", "") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace (" ", "") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace (" ", "") for i in x])
```

```python
movies.head()
```

```python
#make new col -> Tags 
#And in tags concatenate overview	genres	keywords	cast	crew this cols 
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#we don't want these cols -> overview, genres, keywords, cast, crew 
#so we will remove thats cols and keep the rest of the cols 
#we will make new data frame called new_df 
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
#Now the tags col is in list we will convert it into string 
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#Converting into lower case

new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
       y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
